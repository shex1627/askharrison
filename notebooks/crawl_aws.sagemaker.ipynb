```python
from urllib.parse import urljoin
from urllib.parse import urlparse, urlunparse

def generate_complete_urls(base_url, hrefs):
    """
    Given a base URL and a list of hrefs, generate a list of complete URLs.
    """
    complete_urls = []

    for href in hrefs:
        try:
            # Ignore in-page links
            if href == "#":
                continue
            
            # Join the base_url with the href to create a complete URL
            complete_url = urljoin(base_url, href)
            complete_urls.append(complete_url)
        except:
            # Ignore exceptions
            pass
    return complete_urls


def remove_duplicate_urls(urls):
    """
    Remove duplicate URLs and normalize them by stripping fragments.
    """
    unique_urls = set()
    normalized_urls = []

    for url in urls:
        # Parse the URL and remove the fragment
        parsed_url = urlparse(url)
        normalized_url = urlunparse((parsed_url.scheme, parsed_url.netloc, parsed_url.path,
                                     parsed_url.params, parsed_url.query, ""))

        # Add to the list of unique URLs if not seen before
        if normalized_url not in unique_urls:
            unique_urls.add(normalized_url)
            normalized_urls.append(normalized_url)
    
    return normalized_urls
```
