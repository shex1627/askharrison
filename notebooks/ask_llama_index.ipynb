```python
import chromadb
from llama_index import VectorStoreIndex, SimpleDirectoryReader
from llama_index.vector_stores import ChromaVectorStore
from llama_index.storage.storage_context import StorageContext


# initialize client, setting path to save data
db = chromadb.PersistentClient(path="../data/chroma_db")

```

```python
db.list_collections()
```

```python
# create collection
chroma_collection = db.get_or_create_collection("llama_index")
chroma_collection.count()
```

```python
chroma_collection.peek()
```

```python
from llama_index.vector_stores import ChromaVectorStore

# create vector store
chroma_vector_store = ChromaVectorStore(chroma_collection)
index = VectorStoreIndex.from_vector_store(vector_store=chroma_vector_store )
```

```python
llama_index_query_engine = index.as_query_engine()
```

```python
llama_index_query_engine.query("what is llama index")
```
