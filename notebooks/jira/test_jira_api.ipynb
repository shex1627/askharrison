```python
import base64

def encode_credentials(user_email, api_token, encoding='utf-8'):
    text = f'{user_email}:{api_token}'
    bytes_text = text.encode(encoding)
    encoded_text = base64.b64encode(bytes_text).decode(encoding)
    return encoded_text
```

```python
def create_issue(fields, url="https://ftdalpha.atlassian.net/rest/api/3/issue", auth=auth):
    headers = {
    "Accept": "application/json",
    "Content-Type": "application/json"
    }
    payload = json.dumps({"fields": fields})
    response = requests.request("POST", url, data=payload, headers=headers, auth=auth)
    return response
```
